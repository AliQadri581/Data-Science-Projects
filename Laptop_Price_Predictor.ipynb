```python
laptops['Memory'] = laptops['Memory'].astype(str).replace('\.0', '', regex=True) #Memory column ki values ko string mein convert kiya gaya hai aur decimal point ke baad ke zeros ko remove kiya gaya hai.
laptops["Memory"] = laptops["Memory"].str.replace('GB', '')
laptops["Memory"] = laptops["Memory"].str.replace('TB', '000')
new = laptops["Memory"].str.split("+", n = 1, expand = True) #Memory column ko '+' ke basis par split kiya gaya hai aur do alag columns mein store kiya gaya hai.

laptops["first"]= new[0] #Split kiye gaye columns ko 'first' aur 'second' columns mein assign kiya gaya hai.
laptops["first"]=laptops["first"].str.strip()

laptops["second"]= new[1]

laptops["Layer1HDD"] = laptops["first"].apply(lambda x: 1 if "HDD" in x else 0)
laptops["Layer1SSD"] = laptops["first"].apply(lambda x: 1 if "SSD" in x else 0)
laptops["Layer1Hybrid"] = laptops["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
laptops["Layer1Flash_Storage"] = laptops["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

laptops['first'] = laptops['first'].str.replace(r'\D', '') #'first' column se non-numeric characters ko remove kiya gaya hai.

laptops["second"].fillna("0", inplace = True) # 'second' column mein jo NaN values hain unhe zero se fill kiya gaya hai.

laptops["Layer2HDD"] = laptops["second"].apply(lambda x: 1 if "HDD" in x else 0)
laptops["Layer2SSD"] = laptops["second"].apply(lambda x: 1 if "SSD" in x else 0)
laptops["Layer2Hybrid"] = laptops["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
laptops["Layer2Flash_Storage"] = laptops["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

laptops['second'] = laptops['second'].str.replace(r'\D', '')

laptops["first"] = laptops["first"].astype(int)
laptops["second"] = laptops["second"].astype(int)

laptops["HDD"]=(laptops["first"]*laptops["Layer1HDD"]+laptops["second"]*laptops["Layer2HDD"])
laptops["SSD"]=(laptops["first"]*laptops["Layer1SSD"]+laptops["second"]*laptops["Layer2SSD"])
laptops["Hybrid"]=(laptops["first"]*laptops["Layer1Hybrid"]+laptops["second"]*laptops["Layer2Hybrid"])
laptops["Flash_Storage"]=(laptops["first"]*laptops["Layer1Flash_Storage"]+laptops["second"]*laptops["Layer2Flash_Storage"])#These lines calculate the total storage for different types ('HDD', 'SSD', 'Hybrid', 'Flash Storage')

laptops.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
X = laptops.drop(columns=['Price'])
```

```python
Y = laptops['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.15,random_state=2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
```

```python
first = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')
#Onehotencoder changes the catagorical values to binary which is understandable to ML algos
#column tranformer is transforming the columns to apply Onehotendcoder to the named columns rest will be the same
#Column transformer seperates the numerical values and categorical features
```

```python
second = DecisionTreeRegressor(max_depth=8)

pipe = Pipeline([
    ('step1',first),
    ('step2',second)
])  #raw cleaned data is given to CT and ML algo, will provide prediction further
#pipeline arranges the data processing sequence.

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('Mean',mean_absolute_error(y_test,y_pred))
```
