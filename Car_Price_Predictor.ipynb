```python
X = cars.drop(columns='Price') #droping Price
Y = cars['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
One = OneHotEncoder()
One.fit(X[['name', 'company', 'fuel_type']])
#it changes the catagorical values to binary which is understandable to ML algos
```

```python
CT = make_column_transformer((OneHotEncoder(categories = One.categories_),['name', 'company', 'fuel_type']),remainder='passthrough')
#transforming the columns to apply Onehotendcoder to the named columns rest will be the same
```

```python
lr = LinearRegression()
```

```python
pipeline = make_pipeline(CT,lr)
#raw cleaned data is given to CT and Lr will provide prediction further
```

```python
pipeline.fit(X_train,Y_train)
```

```python
Y_prediction = pipeline.predict(X_test)
```

```python
Y_prediction
```

```python
r2_score(Y_test,Y_prediction)
```

```python
scores=[]  #i have done this cuz there was a lot of fluctuations in predictions so i am taking the max r2 score from the dataset
for i in range(1000):
  X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=i)
  lr = LinearRegression()
  pipeline = make_pipeline(CT,lr)
  pipeline.fit(X_train,Y_train)
  Y_prediction = pipeline.predict(X_test)
  scores.append(r2_score(Y_test,Y_prediction))
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipeline = make_pipeline(CT,lr)
pipeline.fit(X_train,Y_train)
Y_prediction = pipeline.predict(X_test)
r2_score(Y_test,Y_prediction)
```
